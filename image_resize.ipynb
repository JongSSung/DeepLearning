```python
import random
import numpy as np
import os
import cv2
import glob
from PIL import Image
import PIL.ImageOps    
    #       .
num_augmented_images = 1000
file_path = './Data/ugap/'
file_names = os.listdir(file_path)
total_origin_image_num = len(file_names)
augment_cnt = 1

for i in range(1, num_augmented_images):
    change_picture_index = random.randrange(1, total_origin_image_num-1)
    print(change_picture_index)
    print(file_names[change_picture_index])
    file_name = file_names[change_picture_index]
    
    origin_image_path = file_path + file_name
    print(origin_image_path)
    image = Image.open(origin_image_path)
    random_augment = random.randrange(1,4)
    
    if(random_augment == 1):
        #  
        print("invert")
        inverted_image = image.transpose(Image.FLIP_LEFT_RIGHT)
        inverted_image.save(file_path + 'inverted_' + str(augment_cnt) + '.jpg')
        
    elif(random_augment == 2):
        # 
        print("rotate")
        rotated_image = image.rotate(random.randrange(-20, 20))
        rotated_image.save(file_path + 'rotated_' + str(augment_cnt) + '.jpg')
        
    augment_cnt += 1
    
```

```python
import time
import os
from PIL import Image
import sys
root_dir = './Data/Prac/' # 
 
img_path_list = []
possible_img_extension = ['.jpg', '.jpeg', '.JPG', '.bmp', '.png'] #  
 
for (root, dirs, files) in os.walk(root_dir):
    if len(files) > 0:
        for file_name in files:
            if os.path.splitext(file_name)[1] in possible_img_extension:
                img_path = root + '/' + file_name
                
                #  \  / 
                img_path = img_path.replace('\\', '/') # \ \\          
                img_path_list.append(img_path)
                             
print(img_path_list)

print("Process Start.")

start_time = time.time()

 #    
COUNT = 1

for a in range(len(img_path_list)):

    image_filename = img_path_list[a]

    #    
    out_dir ="augmentation"
    if out_dir not in os.listdir():
        os.mkdir(out_dir)

    #    
    image = Image.open(image_filename)
    Xdim, Ydim = image.size

    #  
    #    .
    temp_new_file_name = "%06d_77.jpg" %COUNT
    #  1 .
    COUNT += 1
    #   .
    image.save(out_dir + "/" + temp_new_file_name)
    image.close()

    #     .
    FILELIST = [temp_new_file_name]

    #       . 2 1
    for i in range(len(FILELIST)):
        #  .
        image = Image.open(out_dir + "/" + FILELIST[i])
        #       .
        new_temp_name = "%06d_77.jpg" %COUNT
        #     count 1 .
        COUNT += 1
        #   .
        image = image.transpose(Image.FLIP_LEFT_RIGHT)
        #    .
        image.save(out_dir + "/" + new_temp_name)
        image.close()
        #    .
        FILELIST.append(new_temp_name)

    #       . 2 2
    for i in range(len(FILELIST)):
        #  .
        image = Image.open(out_dir + "/" + FILELIST[i])
        #       .
        new_temp_name = "%06d_77.jpg" % COUNT
        #     count 1 .
        COUNT += 1
        #   .
        image = image.transpose(Image.FLIP_TOP_BOTTOM)
        #    .
        image.save(out_dir + "/" + new_temp_name)
        image.close()
        #    .
        FILELIST.append(new_temp_name)

    #      1 . 2 3 * 180
    for el in FILELIST:
        for i in range(180):
            #  1,000 .
            #  1000   .
            if i > 180:
                break
            #  .
            image = Image.open(out_dir + "/" + el)
            #       .
            new_temp_name = "%06d_77.jpg" % COUNT
            #     count 1 .
            COUNT += 1
            #  .
            image = image.rotate(i+1)
            #       resize() .
            image = image.resize((Xdim, Ydim))
            #    .
            image.save(out_dir + "/" + new_temp_name)
            image.close()

    print("Process Done.")

    end_time = time.time()
    print("The Job Took " + str(end_time - start_time) + " seconds.")
```
